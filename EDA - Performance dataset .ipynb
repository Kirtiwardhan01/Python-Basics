```python
#Importing relevant libraries and modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Reading the dataset 

```python
df = pd.read_csv(r"C:\Users\hp\Desktop\kirtiwardhan\Datasets\StudentsPerformance.csv")
df.head()
```

# Pairplot plots pairwise relationships in a dataset
# We can draw scatterplots for joint relationships and histograms for univariate distributions

```python
sns.pairplot(df)
```

# We can use kernel density estimates for univariate plots

```python
sns.pairplot(df, diag_kind="kde")
```

# In the same way we can also pairplot with any two features

```python
sns.pairplot(df, height=3,
                  vars=["math score", "reading score"])
```

# We can fit linear regression models to the scatter plots

```python
sns.pairplot(df, kind="reg")
```

# Using matplot.pyplot library we can visualize histogram

```python
df['math score'].hist()
```

```python
df['reading score'].hist()
```

```python
sns.distplot(df['writing score'])
```

# The boxplot helps in showing the quartiles of the dataset and helps in identfying any outliers 

```python
sns.boxplot(data=df, orient="h", palette="Set2")
```
